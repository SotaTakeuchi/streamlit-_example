他の企業の株価も追加できるように汎用化させる

```python
days = 20

Tickers = {
    'apple': 'AAPL',
    'facebook' : 'FB',
    'google': 'GOOGL',
    'microsoft':'MSFT',
    'netfilx':'NFLX',
    'amazon':'AMZN'
}
#企業名がkeyになっている。Tickers.keyで確認可能

def get_data(days,tickers):
    df = pd.DataFrame()
    for company in Tickers.keys():
        tkr = yf.Ticker(Tickers[company])
        hist = tkr.history(period=f'{days}d')
        hist.index = hist.index.strftime('%d %B %Y')
        hist = hist[['Close']]
        hist.columns = [company]
        hist = hist.T
        hist.index.name = 'Name'
        df = pd.concat([df, hist])
    return df
```

```python
get_data(20, Tickers)
```

```python
import altair as alt
```

```python
df
```

```python
companies = ['apple', 'facebook']
data = df.loc[companies]#ilocにすると番号で指定、locだとカラム名で指定可能
data
```

```python
data.sort_index()
```

altairはpandasのデータ型をそのまま引き継ぐことが可能

```python
#横軸に日付が必要
data=data.T.reset_index()
data.head()
#これでindex部分からDataの取り除きに成功
```

しかし本当はDataに対して「日付、企業1の株価、企業2の株価」という紐づけかたではなくて  
Dataに対して、「日付、企業名、株価」にしたい

```python
data = pd.melt(data, id_vars=['Date']).rename(columns = {'value' : 'Stock Prices(USD)'})
data
#上のセルで綺麗に成形されたデータをわざと崩す(=melt)
#なにを基準に崩すのか = Date
#ピポットテーブル
```

これで「日付、企業名、株価」の取得が可能になった

```python
#グラフの可視化
chart = (
    alt.Chart(data)
    .mark_line(opacity = 0.8)#折れ線グラフ
    .encode(#軸
        x = "Date:T",#Tはタイム。スペース開けてはいけない
        y = alt.Y("Stock Prices(USD):Q", stack = None),#Q:定量化
        color = 'Name:N'
    )
)
chart
```

いくつもの企業名の株価を観測するので、これでは特定の企業の株価が見えづらい  
そこで縦軸の株価の値を自由に調節できるようにしたい

```python
chart = (
    alt.Chart(data)
    .mark_line(opacity = 0.8)#折れ線グラフ
    .encode(#軸
        x = "Date:T",#Tはタイム。スペース開けてはいけない
        y = alt.Y("Stock Prices(USD):Q", stack = None, scale = alt.Scale(domain = [300, 350])),
        color = 'Name:N'
    )
)
chart
```

しかしこれだと御覧の通り、facebookの株価が変な位置に移動してしまう。なので

```python
chart = (
    alt.Chart(data)
    .mark_line(opacity = 0.8, clip = True)#折れ線グラフ
    .encode(#軸
        x = "Date:T",#Tはタイム。スペース開けてはいけない
        y = alt.Y("Stock Prices(USD):Q", stack = None, scale = alt.Scale(domain = [300, 350])),
        color = 'Name:N'
    )
)
chart
```
